## Exploratory Data Analysis

```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 

from tqdm import tqdm 
tqdm.pandas()

import nltk

train_dset_df = pd.read_csv("./dataset/quora/train.csv")
test_dset_df = pd.read_csv("./dataset/quora/test.csv")
```

```python
train_dset_df.drop(inplace=True, axis="columns", labels="qid")
```

```python
grouped_train_dset_df = train_dset_df.groupby(by="target")
```

```python
len(train_dset_df)
```

```python
zeros_dset = grouped_train_dset_df.get_group(0)
```

```python
from nltk.tokenize import WhitespaceTokenizer
```

```python
wstk = WhitespaceTokenizer()
```

```python
train_dset_df["tokenized"] = train_dset_df["question_text"].progress_apply(wstk.tokenize)
```

```python
train_dset_df["len"] = train_dset_df["tokenized"].progress_apply(len)
```
